```python
corpus = [
    "Countries are focusing on renewable energy and carbon reduction to combat climate change.",
    "AI is transforming industries, replacing some jobs while creating new opportunities.",
    "The race heats up, with key issues like healthcare, security, and voter turnout.",
    "The pandemic caused global economic disruption, with a focus on recovery plans and stimulus.",
    "Territorial disputes are escalating, with rising military presence and diplomatic strain.",
    "Social mediaâ€™s impact on mental health is mixed, prompting calls for regulation.",
    "Conflicts and climate change are driving a global migration surge, overwhelming refugee camps.",
    "NASA and SpaceX work towards Mars missions, focusing on sustainability and life support.",
    "The pandemic caused global economic disruption, with a focus on recovery plans and stimulus.",
    "Quantum computing could revolutionize fields like cryptography and medicine, though scalability is an issue.",
    "The EV market is growing, but challenges like charging infrastructure and battery production remain."

]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(corpus)
print("Embedding shape:",embeddings.shape)
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
print("Number of vectors indexed:", index.ntotal)
```

```python
query = "climate change"
query_vec = model.encode([query])

k = 3
distances, indices = index.search(np.array(query_vec), k)

print("query:",query)
for i, idx in enumerate(indices[0]):
  print(f"{i+1}. {corpus[idx]} (distances = {distances[0][i]:.4f})")
```
